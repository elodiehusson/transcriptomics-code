#### **Data Selection**

We decided to keep 60.000 cells of AD, PD, and control patients. They were randomly selected.

```python
## creating masks for AD, PD and control
# case : ignoring majuscules et minuscules
# na : les NaN deviennent false
# values : transform the values into a numpy array
is_AD = adata.obs['disease'].str.contains("Alzheimer", case=False, na=False).values
is_PD = adata.obs['disease'].str.contains("Parkinson", case=False, na=False).values
is_CTRL = adata.obs['disease'].str.contains("normal", case=False, na=False).values

# gets the index of each mask (gets the positions in the AnnData object)
AD_idx = np.flatnonzero(is_AD)
PD_idx = np.flatnonzero(is_PD)
CTRL_idx = np.flatnonzero(is_CTRL)

# random mix of indices
np.random.shuffle(AD_idx)
np.random.shuffle(PD_idx)
np.random.shuffle(CTRL_idx)

# gets 20.000 random indices for each category (it takes the 20.000 first indices of the list, but since it's already randomized, it's random indices)
# it's also important to mix at this point because the dataset is originally ordened by sample or idk what. Anyway, there's a bias there
# we do this to avoid having cells from the same patient
AD_sel = AD_idx[:20000]
PD_sel = PD_idx[:20000]
CTRL_sel = CTRL_idx[:20000]

# list of all the indices to keep to have 60.000 cells
final_idx = np.concatenate([AD_sel, PD_sel, CTRL_sel])
```

```python
# keeps all the genes (columns), but the cells we chose (from the indices)
adata_small = sc.read_h5ad("adata.h5ad")[final_idx, :].copy()
```
